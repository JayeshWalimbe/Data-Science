```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv("Universities .csv")
```

```python
df
```

```python
df.head()
```

```python
df=df.iloc[:,1:]
```

```python
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaled_df=StandardScaler().fit_transform(df)
```

```python
scaled_df
```

```python
from sklearn.cluster import KMeans  # intilizaed the k means 
```

```python
k_c=KMeans(4,random_state=4)
```

```python
k_c.fit_predict(scaled_df)
```

```python
df['Cluster_no']=pd.DataFrame(k_c.fit_predict(scaled_df),columns=['Cluster_no'])
```

```python
df
```

```python
sns.scatterplot(x=df['SAT'],y=df['Expenses'],hue=df['Cluster_no'],palette='Set1')
```

# DENSITY BASED SPATIAL CULSTERING FOR APPLICATIONS WITH NOISE

```python
from sklearn.cluster import DBSCAN
```

```python
dbscan=DBSCAN(eps=2,min_samples=10)
dbscan.fit(scaled_df)
```

```python
dbscan.labels_
```

```python
c1=pd.DataFrame(dbscan.labels_,columns=['cluster'])
```

```python
clustered=pd.concat([df,c1],axis=1)
```

```python
clustered
```

```python
sns.scatterplot(x=clustered["SAT"], y=clustered['Expenses'], hue=clustered['cluster'], palette='Set1')
```
