## linear regression: it is machine learning algorithm based on supervised learning 
### It is statistical method that is used for predictive analysis. Linear regression makes predictions for continuous/real for numeric variables such as cost, age, sales, temperature, product price etc 

### Linear Equation: y=mx+c
### Where 
### y-dependent variable
### x-independent variable
### m-slope/gradient/coefficient
### c-intercept
 

```python
import pandas as pd
import seaborn as sns
from sklearn import linear_model
import warnings 
warnings.filterwarnings('ignore') 
```

```python
df=pd.read_csv("Insurance.csv")
```

```python
df
```

```python
sns.lmplot(x='age',y='premium',data=df)
```

```python
reg=linear_model.LinearRegression()         #regression model 
```

```python
reg.fit(df[['age']],df['premium'])          # model is frame with data point for prediction 
```

```python
reg.predict([[21]])       # It will predict the premium for the 21 age.
```

```python
reg.predict([[50]])       #it will preict the premium for the 50 age.
```

```python
reg.coef_                 # it will give the coefficent 
```

```python
reg.intercept_           # it will give the intercept
```
