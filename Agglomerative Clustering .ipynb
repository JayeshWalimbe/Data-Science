# scaling

```python
def norm(i):
    x = (i - i.min()) / (i.max() - i.min())
    return x
```

```python
norm_df1=norm(df1)
```

```python
norm_df1
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scalar=MinMaxScaler()
scaled_df1=scalar.fit_transform(df1)    # same answer like above only the answer is getting in array 
```

```python
print(scaled_df1)
```

```python
df.columns
```

```python
pd.DataFrame(scaled_df1,columns=df.columns[1:])   # converted into df
```

```python
df1.head()
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
hc=AgglomerativeClustering(n_clusters=2,linkage='ward')
```

```python
hc.fit(scaled_df1)
```

```python
dir(hc)
```

```python
hc_pred=hc.fit_predict(scaled_df1)
```

```python
hc_pred
```

```python
pd.DataFrame(hc_pred,columns=["Cluster_no"])
```

```python
df['Cluster_no']=pd.DataFrame(hc_pred,columns=['cluster_no'])
```

```python
sns.scatterplot(x=df['SAT'], y=df['Expenses'],hue=df['Cluster_no'],palette='Set1')
```
